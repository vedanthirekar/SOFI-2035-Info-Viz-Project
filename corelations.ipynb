```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---------- Read Data ----------
# Replace the file path with your actual path if different
df = pd.read_excel("D:\MSDS\Info Viz\Client Project SOFI\excel1.xlsx")

# Inspect the data (optional)
print(df.head())

numeric_df = df.select_dtypes(include=['float64', 'int64'])
if 'Year' in numeric_df.columns:
    numeric_df = numeric_df.drop(columns=['Year'])

# Compute correlation
corr = numeric_df.corr()

# Plot heatmap without annotations
plt.figure(figsize=(8, 6))
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5, cbar_kws={'label': 'Correlation'})
plt.title("Correlation Heatmap of Health Indicators", fontsize=14)
plt.tight_layout()
plt.show()
```

```python


# Extract Year and desired variables
cols = [
    "Year",
    "Life expectancy",
    "Mortality rate, infant",
    "Undernourishment",
    "Health expenditure",
    "Physicians",
    "Drinking water"
]
df = df[cols]

# Plot multiple lines
plt.figure(figsize=(10, 6))
for col in cols[1:]:
    plt.plot(df["Year"], df[col], label=col, linewidth=2)

plt.title("Health & Well-being Indicators (1990â€“2023)", fontsize=14, weight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Normalized Value", fontsize=12)
plt.legend(title="Indicators", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create correlation matrix
corr = df.drop(columns=['Year']).corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create mask for upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Create improved heatmap
sns.heatmap(corr, 
            mask=mask,  # Mask upper triangle
            annot=True,  # Show correlation values
            cmap='RdYlBu_r',  # Red-Yellow-Blue colormap
            fmt='.2f',  # Format annotations to 2 decimal places
            square=True,  # Make cells square
            linewidths=1,  # Add line between cells
            cbar_kws={'label': 'Correlation Coefficient'},
            annot_kws={'size': 10})

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Add title with padding
plt.title('Correlation Heatmap of Health Indicators\n(Lower Triangle)', pad=20, fontsize=14)

# Adjust layout to prevent label cutoff
plt.tight_layout()
plt.show()

# Print correlation analysis
print("\nKey Correlation Findings:")
print("-" * 50)

# Find strong correlations (absolute value > 0.7)
strong_corr = []
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > 0.7:
            strong_corr.append((corr.index[i], corr.columns[j], corr.iloc[i, j]))

# Sort by absolute correlation value
strong_corr.sort(key=lambda x: abs(x[2]), reverse=True)

# Print findings
for var1, var2, corr_val in strong_corr:
    relationship = "positive" if corr_val > 0 else "negative"
    strength = "very strong" if abs(corr_val) > 0.9 else "strong"
    print(f"{var1} and {var2}: {strength} {relationship} correlation (r = {corr_val:.2f})")
```

```python
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns

# # Read Excel file
# df = pd.read_excel("excel1.xlsx")

# # Health-related columns (from your data)
health_vars = [
    "Life expectancy",
    "Mortality rate, infant",
    "Undernourishment",
    "Health expenditure",
    "Physicians",
    "Drinking water"
]

# Plot boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[health_vars], palette="Set3", width=0.6)
plt.title("Outlier Detection in Health Indicators", fontsize=14, weight='bold')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
outliers = {}

for col in health_vars:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1 * IQR
    upper = Q3 + 1 * IQR
    
    mask = (df[col] < lower) | (df[col] > upper)
    outliers[col] = df.loc[mask, ["Year", col]]

# Print outliers
for k, v in outliers.items():
    if not v.empty:
        print(f"\nOutliers in {k}:")
        print(v)
```

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
num_cols = num_cols.drop('Year')

# --- BOX PLOT FOR ALL VARIABLES ---
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols], palette="Set2", width=0.6)
plt.title("Outlier Detection for All Numeric Variables", fontsize=14, weight='bold')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()

# --- OPTIONAL: PRINT OUTLIERS ---
outliers = {}

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    mask = (df[col] < lower) | (df[col] > upper)
    outliers[col] = df.loc[mask, [col]]

# Display summary
for col, vals in outliers.items():
    if not vals.empty:
        print(f"\nOutliers in {col}:")
        print(vals)
```
